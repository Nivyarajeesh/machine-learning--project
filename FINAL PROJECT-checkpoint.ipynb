# Feature selection

```python
# Feature Selection using SelectKBest
k = 12  # Choose the number of top features you want to select
selector = SelectKBest(score_func=f_classif, k=k)
X_selected = selector.fit_transform(x,y)

# Get the selected feature indices
selected_feature_indices = selector.get_support(indices=True)

# Get the column names of the selected features
selected_feature_names = x.columns[selected_feature_indices]


# Plot the bar chart of selected features
selected_feature_scores = selector.scores_[selected_feature_indices]
selected_features_df = pd.DataFrame({
    'Feature': selected_feature_names,
    'Score': selected_feature_scores
})

# Print the scores of the selected features
print("Scores of selected features:")
print(selected_features_df)

# Sort the features by their scores in descending order
selected_features_df = selected_features_df.sort_values(by='Score', ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Score', y='Feature', data=selected_features_df, palette='viridis')
plt.title('Selected Features by SelectKBest Scores')
plt.xlabel('Score')
plt.ylabel('Feature')
plt.show()

```

```python
# Create a DataFrame with the data of the selected features
selected_features_data = pd.DataFrame(x, columns=selected_feature_names)

selected_features_data.shape
```

# skewness

```python
skewness = selected_features_data.skew()
print("Skewness of the data:")
print(skewness)
```

 Findings:- Most features have positive skewness, indicating a right-skewed distribution.

```python
# Apply Transformation to Reduce Skewness
# Apply log transformation to features with skewness greater than 0.5
skewed_features = skewness[skewness > 0.5].index
selected_features_data[skewed_features] = np.log1p(selected_features_data[skewed_features])

# You can try other transformations like square root or Box-Cox transformation

# Step 3: Re-Calculate Skewness
skewness_after_transformation = selected_features_data.skew()

# Print Skewness before and after transformation
print("Skewness before transformation:")
print(skewness)
print("\nSkewness after transformation:")
print(skewness_after_transformation)
```

 #Findings:- The skewness values of the features have been reduced compared to before.However, some features still exhibit high skewness values, such as AcceptedCmps

# outlier detection

```python
# Step 3: Identify outliers using IQR method
Q1 = X_normalized_data.quantile(0.25)
Q3 = X_normalized_data.quantile(0.75)
IQR = Q3 - Q1
outliers = ((X_normalized_data < (Q1 - 1.5 * IQR)) | (X_normalized_data > (Q3 + 1.5 * IQR))).any(axis=1)

# Step 4: Count the number of removed outliers
num_removed_outliers = outliers.sum()
print("Number of removed outliers:", num_removed_outliers)

# Step 5: Remove outliers
X_no_outliers = X_normalized_data[~outliers]
Y_no_outliers = y[~outliers]

# Step 6: Print the shapes of X and Y after removing outliers
print("Original DataFrame shape:", df.shape)
print("Shape of X without outliers:", X_no_outliers.shape)
print("Shape of Y without outliers:", Y_no_outliers.shape)

# Step 7: Plot the boxplot of X without outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=X_no_outliers)
plt.title('Box Plot of Features after Removing Outliers')
plt.xticks(rotation=45)
plt.show()
```

# Random Forest

```python

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialize the Random Forest classifier
rf_model = RandomForestClassifier(random_state=42)

# Train the model with all selected features
rf_model.fit(X_train, y_train)

# Now, let's predict based on only the 'income' feature
# Extract 'income' feature from the original dataset
income_train = X_train[['Income']]
income_test = X_test[['Income']]

# Initialize and train the model with the 'income' feature only
rf_model_income = RandomForestClassifier(random_state=42)
rf_model_income.fit(income_train, y_train)

# Predict responses based on the 'income' feature
income_predictions = rf_model_income.predict(income_test)

# Combine income and prediction results into a DataFrame for easy analysis
income_prediction_df = pd.DataFrame({
    'income': income_test['Income'],
    'predicted_response': income_predictions,
    'actual_response': y_test
})

# Display the DataFrame
print("Income and Predicted Response DataFrame:")
print(income_prediction_df.head())

# Evaluate the accuracy of predictions based on income
income_accuracy = accuracy_score(income_prediction_df['actual_response'], income_prediction_df['predicted_response'])
print(f'Income-based Prediction Accuracy: {income_accuracy}')

# Print classification report
print('Classification Report for Income-based Prediction:')
print(classification_report(income_prediction_df['actual_response'], income_prediction_df['predicted_response']))

# Print confusion matrix
conf_matrix_income = confusion_matrix(income_prediction_df['actual_response'], income_prediction_df['predicted_response'])
print('Confusion Matrix for Income-based Prediction:')
print(conf_matrix_income)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_income, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Income-based Prediction')
plt.show()
```
